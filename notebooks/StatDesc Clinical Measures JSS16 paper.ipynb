This notebook calculates the descriptive statistics of the flexibility and strength measures grouped by intervention group and condition of the Fukuchi et al. (2016) paper. Part of these values were presented in Table 2 of the aforementioned manuscript.

## Abbreviations list
* HADD_FX: Hip Adduction flexibility 
* AGAS_FX: Gastrocnemius flexibility 
* ASOL_FX: Soleus flexibility 
* HER_FX: Hip ER flexibility 
* HIR_FX: Hip IR flexibility 
* HABD_ST: Hip abductors strength
* HEXT_ST: Hip extensors strength
* APF_ST: Ankle PF strength
* STG: Strengthening group
* FXG: Flexibility group
* CTG: Control group
* PRE: Pre-intervention measure
* POS: Post-intervention measure

## Setup environment

```python
import os
import numpy as np
import pandas as pd
```

## Import file with data from each participant included in the study

```python
path2 = '..\data'
filename = os.path.join(path2,'data.csv')
df = pd.read_csv(filename,index_col=0)
```

```python
df.head()
```

```python
df.tail()
```

## Descriptive stats

```python
pd.set_option('precision', 2)
print('Descriptive stats of clinical vars grouped by intervention group and conditions (mean and std):')
df.groupby(['Group','Condition'])[df.columns.tolist()].agg([np.mean, np.std])
```
